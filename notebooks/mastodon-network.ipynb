### Example script to fetch follwers of the user

Another method account_following() fetches users that are followed by the input user.

```python
from mastodon import Mastodon
import pprint
import time


# DON'T FORGET to change path to your access token
mastodon = Mastodon(
    api_base_url="https://mastodon.social",
    access_token="config/pytooter_userscred.secret",
)


user_id = "1"
max_id = None
batch_num = 0
total_num_followrs = 0

# Get user's followers
start_time = time.time()
while True:
    try:
        followers = mastodon.account_followers(
            id=user_id,
            max_id=max_id,
        )
        max_id = followers[-1]["id"]
        followers_list = list(followers)
        num_followers = len(followers)
        total_num_followrs += num_followers
        print(f"Number of retrieved followers: {total_num_followrs}")

        batch_num += 1
        print(f"Number of processed batches: {batch_num}")
        end_time = time.time()
        elapsed_time = end_time - start_time
        print(f"Elapsed time: {elapsed_time}")
        print(50 * "==")
    except Exception as e:
        print(f"Fetching followers exception: {e}")
    if total_num_followrs >= 1000:
        break
pprint.pprint(followers)
```
